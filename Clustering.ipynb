## Using Clustering to group data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans, SpectralClustering
import warnings
```

```python
df = pd.read_csv('data/spirals.csv')
```

```python
data, labels = datasets.make_blobs(n_samples=100, n_features=2, centers=3, random_state = 1)
```

```python
plt.scatter(data[:,0],data[:,1])
plt.show()
```

Data with correct label coloring

```python
plt.scatter(data[:,0],data[:,1], c = labels)
plt.show()
```

Using KMeans to find labels


```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(data)
plt.scatter(data[:,0], data[:,1], c=kmeans.labels_)
plt.show()
```

Note: We can see that the clusterings are correct, but two of the labels are just swapped. This is confirmed below when we print out the labels

```python
print(kmeans.labels_)
print(labels)
```

```python
data, labels = datasets.make_blobs(n_samples=100,n_features=2, centers=[[0,0],[0,0.5],[0,-0.5]], random_state= 1)
plt.scatter(data[:,0],data[:,1])
plt.show()
```

Data with correct label coloring

```python
plt.scatter(data[:,0],data[:,1], c = labels)
plt.show()
```

The visual inspection shows us that there are no clear clusters. It is no surprise that when we use KMeans to find labels, it is highly inaccurate. 

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(data)
plt.scatter(data[:,0], data[:,1], c=kmeans.labels_)
plt.show()
```

```python
data, labels = datasets.make_moons(n_samples=500, noise=.05, random_state = 0)
plt.scatter(data[:,0],data[:,1])
plt.show()
```

Data with correct label coloring

```python
plt.scatter(data[:,0],data[:,1], c = labels)
plt.show()
```

Spectral Clustering with affinity = 'rbf'

```python
clustering = SpectralClustering(n_clusters=2, assign_labels="discretize",random_state=0).fit(data)
plt.scatter(data[:,0], data[:,1], c=clustering.labels_)
plt.show()
```

Spectral Clustering with affinity = 'nearest_neighbors'

```python
clustering = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', assign_labels="discretize",random_state=0).fit(data)
plt.scatter(data[:,0], data[:,1], c=clustering.labels_)
plt.show()
```
